an index value can be duplicated; calling on that index will select all occurences with that label

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)), 
        index=['a', 'c', 'd'], 
        columns=['Ohio', 'Texas', 'California'])
frame
```

```python
frame = frame.reindex(['a', 'b', 'c', 'd'])
frame
```

```python
states = ['Texas', 'Utah', 'California']
frame.reindex(columns=states)
```

```python
frame.reindex(fill_value=789)
```

```python
frame = frame.reindex(['a', 'b', 'c', 'd', 'e'], fill_value=789)
frame
```

slicing a series by the label includes the endpoint

##Indexing options with DataFrame
<br/>
df[val] = select single column or sequence of columns from the DF; special case conveniences: boolean array (filter rows), slice (slice rows), or boolean DF (set values based on some criterion)
<br/>
df.loc[va] = selects single row or subset of rows from the DF by label
<br/>
df.loc[:, val] = selects single column or subset of columns by label
<br/>
df.lov[val1, val2] = sekect bith rows and columns by label
<br/>
df.iloc[where] = selects single row or subset of rows from the DF by integer position
<br/>
df.iloc[:, where] = selects single column or subset of columns by integer position
<br/>
df.iloc[where1, where2] = select both rows and columns by integer position
<br/>
df.at[label1, label2] = select a single scalar value by row and column label
<br/>
df.iat[i, j] = select a single svalar value by row and column position (integers)
<br/>
reindex method = select either row or columns by label
<br/>
get_value, set_value methods = select single value by row and column label
