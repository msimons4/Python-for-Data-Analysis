```python
mnames = ['movie_id', 'title', 'genres']
```

```python
movies = pd.read_table('datasets/movies.dat', sep='::', header=None, names=mnames, engine='python')
```

```python
movies[:10]
```

```python
all_genres = []
for x in movies.genres:
    all_genres.extend(x.split('|'))

genres = pd.unique(all_genres)
genres
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
```

```python
gen = movies.genres[0]
gen
```

```python
gen.split('|')
```

```python
dummies.columns.get_indexer(gen.split('|'))
```

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[9]
```

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
```

```python
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
# looks for [A-Z0-9._%+-] then the @ sign; then these [A-Z0-9.-] followed by a \ or .; then [A-Z]{2,4} for the com/net/gov/etc
```

```python
regex = re.compile(pattern, flags=re.IGNORECASE)
```

```python
regex.findall(text)
```

```python
m = regex.search(text)
m
```

```python
m = regex.search(text,21)
m
```

```python
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex = re.compile(pattern, flags=re.IGNORECASE)
m = regex.match('wesm@bright.net')
m.groups()
```

```python
regex.findall(text)
```
