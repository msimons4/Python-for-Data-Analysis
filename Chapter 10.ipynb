```python
df = pd.DataFrame({'category': ['a', 'a', 'a', 'a', 'b', 'b', 'b', 'b'], 'data': np.random.randn(8), 'weights': np.random.rand(8)})
df
```

```python
grouped = df.groupby('category')
get_wavg = lambda g: np.average(g['data'], weights=g['weights'], returned=True)
grouped.apply(get_wavg)
```

```python
sum_of_weights_a = []
sum_of_weights_b = []
for i in range(0,len(df['category'])):
    if df['category'][i] == 'a':
        sum_of_weights_a.append(df['weights'][i])
    else:
        sum_of_weights_b.append(df['weights'][i])
        
print('sum of a: {} \nsum of b: {}'.format(sum(sum_of_weights_a), sum(sum_of_weights_b)))
```

```python
close_px = pd.read_csv('examples/stock_px_2.csv', parse_dates=True, index_col=0)
close_px.head()
```

```python
spx_corr = lambda x: x.corrwith(x['SPX'])
```

```python
rets = close_px.pct_change().dropna()
```

```python
rets[:10]
```

```python
rets_test = close_px.pct_change(periods=2213).dropna()
rets_test[-10:]
```

```python
close_px.tail()
```

```python
get_year = lambda x: x.year
```

```python
by_year = rets.groupby(get_year)
```

```python
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda y: y['AAPL'].corr(y['MSFT']))
```

```python
def corr_test(x, y):
    return close_px[x].corr(close_px[y])
```

```python
#corr_test('AAPL', 'MSFT')
by_year.apply(corr_test('AAPL', 'MSFT'))
```

```python
import statsmodels.api as sm
```

```python
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year.apply(regress, 'AAPL', ['SPX'])
```
